# DAUM에서 많이 본 랭킹 뉴스

```
import requests 
# Python 기본 모듚 중 하나. 웹 서버에 request를 보내고 response를 받는 기능

from bs4 import BeautifulSoup  # HTML parsing
```

```
url = 'https://news.daum.net/ranking/popular'  # request를 보낼 웹 사이트 주소
```

```
response = requests.get(url)  # 웹 서버에 get 방식의 request을 보냄
#> get() 함수의 리턴 값이 웹 서버가 보내 준 response(응답)

print(response)
#> Response [200]: 요청 처리 성공(Sucess)
```

```
html = response.text.strip()  # Response 객체의 text 속성의 값 - HTML 코드
print(html[:300])
print(html[-100:])
```

```
# HTML을 분석하기 위한 BeautifulSoup 객체를 생성
soup = BeautifulSoup(markup=html, features='html5lib')
```

```
# 웹 브라우저의 개발자 도구를 사용해서
# HTML 페이지에서 원하는 element를 찾을 수 있는 CSS selector를 찾음
len(soup.select('a.link_txt'))
```

# DAUM 뉴스 랭킹 댓글 많은 뉴스 1~20위까지 뉴스의 순위, 제목, 댓글수를 출력

```
url3 = 'https://news.daum.net/ranking/bestreply'
```

```
response3 = requests.get(url3)   # 서버에 요청을 보냄
print(response3)
html3 = response3.text.strip()    # 응답에서 HTML 코드를 추출
soup = BeautifulSoup(markup=html3, features='html5lib')   # BS 객체를 생성
```

```
title_selector = 'ul.list_news2 > li:nth-of-type({i}) > div.cont_thumb > strong > a.link_txt'
```

```
reply_selector = 'ul.list_news2 > li:nth-of-type(3) > div.rank_num > span.info_rank > span.ico_news2'
print(soup.select_one(reply_selector))
```

```
for i in range(1, 21):   # 1~20위까지 
    title_selector = f'ul.list_news2 > li:nth-of-type({i}) > div.cont_thumb > strong > a.link_txt'
    reply_selector = f'ul.list_news2 > li:nth-of-type({i}) > div.rank_num > span.info_rank > span.ico_news2'
    title = soup.select_one(title_selector).text.strip()
    reply = soup.select_one(reply_selector).text.strip()
    print(f'[{i}]', reply, title)
```

# DAUM 뉴스 랭킹 연령별 뉴스에서 연령별/성별 5위까지 뉴스의 제목 링크 URL을 출력

```
url_age_gen = 'https://news.daum.net/ranking/age'   # request를 보낼 URL
response4 = requests.get(url_age_gen)               # 서버로 request를 보내고, response를 받음
print(response4)
```

```
html4 = response4.text.strip()                       # 서버로부터 받은 response에서 HTML 코드를 추출
soup = BeautifulSoup(markup=html4, features='html5lib') # HTML을 분석할 수 있는 

# 태그 ul : unordered list
# 태그 ol : ordered list
# 태그 li : list item
```

```
selector = 'div.item_age.item_20s > div.rank_female > ol > li.num_news.num1 > a.link_txt'
soup.select_one(selector)
```

```
selector = 'div.item_age.item_20s > div.rank_female > ol a.link_txt'
soup.select(selector)
```

```
# div.클래스1.클래스2. 인 경우 클래스 둘 중 하나만 선택 가능함

selector = 'div.item_age.item_20s > div.rank_male > ol a.link_txt'
soup.select(selector)

selector = 'div.item_30s > div.rank_female > ol a.link_txt'
soup.select(selector)
```

```
ages = range(20, 60, 10)      # 연령대: 20, 30, 40, 50대
genders = ('female', 'male')  # 성별

for age in ages:
    for gender in genders:
        category_selector = f'div.item_age.item_{age}s > div.rank_{gender} > strong.txt_news'
        category = soup.select_one(category_selector).text.strip()
        print('---', category, '---')

        selector = f'div.item_{age}s > div.rank_{gender} > ol a.link_txt'
        news_top5 = soup.select(selector)
        for link in news_top5:
            print(link['href'], link.text)
```

```
for a in range(2, 6):
    print(f'# {a}0대 여성 Top5  >>>>>')
    for i in range(1, 6):
        selector = f'div.item_age.item_{a}0s > div.rank_female > ol > li:nth-of-type({i}) > a.link_txt'
        link = soup.select_one(selector)
        print(f'[{i}] {link.text}')
        print(link.get('href'))
    print('\n')
    print(f'# {a}0대 남성 Top5  >>>>>')
    for i in range(1, 6):
        selector = f'div.item_age.item_{a}0s > div.rank_male > ol > li:nth-of-type({i}) > a.link_txt'
        link = soup.select_one(selector)
        print(f'[{i}] {link.text}')
        print(link.get('href'))
    print('\n')
```
