# list comprehension

1 이상 10 이하의 정수들을 리스트에 추가

```
numbers = []
for x in range(1, 11):
    numbers.append(x)

print(numbers)
```

```
numbers2 = [x for x in range(1, 11)] # list comprehension
print(numbers2)
```

10 이하의 양의 정수들 중에서 짝수들을 리스트에 추가

```
# for loop + append()
evens = []
for x in range(1, 11):
    if x % 2 == 0:
        evens.append(x)

print(evens)
```

```
evens = [n for n in range(2, 11, 2)]
print(evens)
```

```
# list comprehension
evens2 = [x for x in range(1, 11) if x % 2 == 0]
print(evens2)
```

```
evens = []
for n in range(1, 6):
    evens.append(n * 2)
print(evens)
```

```
evens = [n * 2 for n in range(1, 6)]
print(evens)
```


1 ~ 5 숫자들의 제곱을 리스트에 추가 [1, 4, 9, 16, 25]

```
# for loop
sqr = [] 
for n in range(1, 6):
    sqr.append(n ** 2)
print(sqr)
```

```
# list comprehension
sqr = [n ** 2 for n in range(1, 6)]
print(sqr)
```

0이상 100 이하인 정수 난수 10개를 리스트에 저장

```
import random
```

```
random.seed(1)

rands = []
for x in range(10):
    rands.append(random.randrange(0, 101))

print(rands)
```

```
random.seed(1)

rands = []
rands = [random.randrange(0, 101) for x in range(10)]
print(rands)
```

# if 문장이 list comprehension 함께 사용되는 경우

1 ~ 10까지 정수들 중에서 짝수들을 리스트에 추가

```
evens = []
for x in range(1, 11):
    if x % 2== 0:
        evens.append(x)

print(evens)
```

```
evens = [x for x in range(1, 11) if x % 2 == 0]
print(evens)
```

홀수들로만 이루어진 새로운 리스트를 만들고 출력

```
print(rands)
odd_rands = [x for x in rands if x % 2]
# if 문에서 숫자가오는 경우 0이 오면 False, 그외에 True
print(odd_rands)
```

10개의 난수를 저장하고 있는 리스트 rands의 값이 짝수이면 True, 홀수이면 False 저장하는 새로운 리스트를 생성 

```
print(rands)
```

```
is_even = []
for x in rands:
    if x % 2 == 0:
        is_even.append(True)
    else:
        is_even.append(False)
print(is_even)
```

```
is_even = [True if x % 2 == 0 else False for x in rands]
# append 할 값이 둘중의 하나이기 때문에 떨어뜨릴 수가 없음. 그래서 for 가 뒤로 감
print(is_even)
```

```
random.seed(42)

# 0 또는 1을 무작위로 10개 저장하는 리스트 gender1을 생성
gender1 = [random.randrange(0, 2) for _ in range(10)]
print(gender1)

# gender1의 원소의 값이 0이면 'male' 1이면 'female'을 갖는 리스트 gender2를 생성
gender2 = ['male' if x == 0 else 'female' for x in gender1]
print(gender2)
```

```
gender2 =[]
for x in gender1:
    if x==0:
        gender2.append('male')
    else:
        gender2.append('female')

print(gender2)
```

```
random.seed(42)

# 0, 1, 2를무작위로 10개를 저장하는 리스트 gender1을 생성
gender1 = [random.randrange(0, 3) for _ in range(10)]
print(gender1)

# gender1의 원소의 값이 0이면 'male', 1이면 'female', 2이면 'unknown'를 저장하는 리스트 gender2를 생성
gender2 = ['male' if x == 0 else 'female' if x == 1 else 'unkown' for x in gender1]
print(gender2)
```

```
gender2 = []
for x in gender1:
    if x == 0:
        gender2.append('male')
    elif x == 1:
        gender2.append('female')
    else:
        gender2.append('unknown')

print(gender2)
```

```
random.seed(123)

# 0 이상 100이하인 정수 난수 10개를 갖는 리스트 scores를 생성
scores = [random.randrange(0, 101) for _ in range(10)]
print(scores)

mean_score = sum(scores) / len(scores)
print(f'평균: {mean_score}')

# scores의 원소들 중 평균보다 큰 숫자들로 이루어진 리스트 above_mean을 생성
above_mean = [x for x in scores if x > mean_score]
# above_mean = [x for x in scores if x > sum(scores) / len(scores)]
print(above_mean)
```

```
words = ['apple', 'banana', 'cherry', 'dragon fruit', 'elderberry']

# words의 단어들의 문자의 개수(문자열의 길이)를 저장하는 리스트 word_length를 생성
word_length = [len(x) for x in words]
print(word_length)
```
