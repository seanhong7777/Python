```
# 난수를 사용하기 위해 필요한 모듈을 import
import random
```

```
# 빈 리스트(empty list) 생성
scores = list()
# scores = []
print(scores)
```

```
# 리스트 scores에 0 이상 100 이하의 정수 난수 5개를 추가
# list.append() 메서드 사용. 

for _ in range(5):
    # s = random.randrange(0, 101)
    # scores.append(s)
    scores.append(random.randrange(0, 101))

print(scores)
```

```
# 리스트 scores에 저장된 값들의 합계 (총점)
total = 0   # 총점을 저장할 변수 

for s in scores:
    total += s

print(f'총점 = {total}')
```

```
# 리스트 scores에 저장된 값들의 평균
average = total / len(scores)

print(f'평균 = {average}')
```

```
# 리스트 scores에 저장된 값들의 분산(variance)
# 각 점수에서 평균을 뺀 값의 제곱들의 평균 

# 각 점수에서 평균을 뺀 값을 제곱해서 누적해 나갈 변수 
sum_of_squares = 0 
for s in scores:
    sum_of_squares += (s - average) ** 2

variance = sum_of_squares / len(scores)
print(f'분산 = {variance}')
```

```
import math
```

```
# 표준편차(standard deviation) = sqrt(분산)
std_dev = math.sqrt(variance)
print(f'표준편차 = {std_dev}')
```

```
# 리스트 scores의 값들 중에서 최댓값 찾기
# 1. 리스트의 첫번째 값을 최댓값이라고 가정
# 2. 리스트의 원소들을 하나씩 꺼내면서
# 3. 리스트의 원소가 가정한 최댓값보다 더 크다면, 찾은 값을 최댓값으로 저장
# 2~3 과정을 리스트의 전체 원소들에 대해서 반복 
print(scores)
max_score = scores[0]
for s in scores:
    if s > max_score:
        max_score = s
    else:
        continue

print(f'최댓값 = {max_score}')
```

```
# 리스트 중에서 최솟값 찾기
print(scores)
min_score = scores[0]
for s in scores:
    if s < min_score:
        min_score = s
    else:
        continue

print(f'최솟값 = {min_score}')
```

```
# 파이썬 기본 함수들
print('sum:', sum(scores))
```

```
# 리스트 정렬(sorting)
# sorted() 함수: 정렬된 새로운 리스트를 반환(return). 원본 리스트는 순서가 바뀌지 않음

asc_sort = sorted(scores)
print(asc_sort)

desc_sort = sorted(scores, reverse=True)
print(desc_sort)

print(scores)   # sorted 함수 호출 후에도 리스트 scores의 순서는 그대로 유지 
```

```
# list.sort() 메서드: 원본 리스트를 정렬해서 바꿈
scores.sort()
print(scores)

scores.sort(reverse=True)
print(scores)
```

```
# Fibonacci Sequence(피보나치 수열)

fibo = [1, 1]
# for loop과 append 메서드를 사용해서 원소개수가 20개인 피보나치 수열을 완성
for i in range(2, 20):
    fibo.append(fibo[i-1] + fibo[i-2])

print(fibo)
```

```
# fibo[i+1] / fibo[i] 비율 19개를 모두 출력
fibo_series = []
for i in fibo:
    if i < len(fibo)-1:
        fibo_series.append(fibo[i+1] / fibo[i])

print(fibo_series)
```

```
# fibo[i+1] / fibo[i] 비율 19개를 모두 출력
golden_ratio = []
for i in range(19):
    golden_ratio.append(fibo[i+1] / fibo[i])

print(golden_ratio)
```
