# 리스트와 반복문

```
scores = [1, 3, 2, 4, 5]
for s in scores:
    print(s)
```

```
for i in range(len(scores)):  # for i in range(5): 
    print(scores[i])
```

```
# enumerate(list): list의 인덱스와 해당 인덱스의 값을 함께 반환
for i, x in enumerate(scores):
    print(i, ':', x)
```

```
# 2차원 리스트: 1차원 리스트들을 원소로 갖는 리스트 
matrix = [
           [1, 2, 3],
           [4, 5, 6],
           [7, 8, 9]
]
print(matrix)
```

```
print(len(matrix))
```

```
matrix[0]
```

```
matrix[0][0]
```

```
# 2차원 리스트와 for loop
for row in matrix:
    print(row)
```

```
for row in matrix:
    for x in row:
        print(x, end = '\t')
    print()
```

```
matrix = [
          [1, 2, 3],
          [4, 5, 6, 7],
          [8, 9]
]
print(matrix)
```

```
# 6 
print(matrix[1][2])
```

```
for row in matrix:
    for x in row:
        print(x, end = '\t')
    print()
```

```
for i in range(len(matrix)):
    for j in range(len(matrix[i])):
        print(matrix[i][j], end = '\t')
    print()
```

```
for i, row in enumerate(matrix):
    for j, x in enumerate(row):
        print(f'matrix[{i}][{j}] = {x}')
```
