# Python 데이터 타입 (Data Type)



*   숫자 타입:
  * int: 정수
  * float: 실수
*   논리 타입
*   문자열 타입(str)
*   NonType: 값이 없음을 나타내는 데이터 타입. None



```
# 변수 n에 정수 100을 저장
n = 100 
print(n)
print(type(n))  # type(x): 값, 변수  x의 데이터 타입을 반환
```

```
pi = 3.141592  # 대입문(assignment statement). 할당문 
print(pi) 
print(type(pi))
```

```
s1 = 'hello'
s2 = '안녕'
print(s1, type(s1))
print(s2, type(s2))
# Python은 문자열을 표기할 때 큰따옴표("") 또는 작은따옴표('')를 사용함 
```

## 명시적 타입 변환(explicit type conversion)

* int(값): 값 -> 정수. 정수로 변환될 수 없는 값이 argument로 전달되면 에러가 발생함
* float(값): 값 -> 실수
* str(값): 값 -> 문자열

```
age = '16'
print(type(age))
```

```
age2 = int(16.5)
print(age2, type(age2))
```

```
pi = float('3.141592')
print(pi, type(pi))
```

## 산술 연산자: `+, -, *, /, **, //, %`

* 산술 연산자들은 숫자 타입(int, float)의 값 또는 변수에서 사용할 수 있음 

```
n1 = int(input('정수를 입력하세요>>> '))  # 입력받은 값을 변수 n1에 저장. n1의 타입은 str
n2 = int(input('정수를 입력하세요>>> '))
print(n1 - n2)
```

* 문자열(str)타입에서는 `+, *`가 사용될 수 있음
* '+': concatenate. 문자열 이어 붙이기
* '*': replicate. 문자열 복제

```
print('Hello' + 'Python')  # 문자열 + 문자열
```

```
print('Hello' * 5)  # 문자열 * 정수 
```

* 논리(bool) 타입: True/False를 저장하는 타입
  * bool 타입의 변수들을 산술 연산을 하면, True는 1로, False는 0으로 변환된 후 산술 연산이 수행됨 
  * 암시적 타입 변환(implicit type conversion): 개발자가 int()와 같은 함수를 사용해서 명시적으로 타입 변환을 하지 않아도, 프로그램 언어에서 자동으로 타입 변환을 수행하는 것 

```
print(type(True))
```

```
# implicity type conversion 
print(True + True) # True True -> 1 + 1 -> 2 
print(True + False)
print(True * True)
print(True * False)
```

* bool 타입 값, 변수들과 사용되는 연산자들 
  * **비교연산자**: `==, !=, >, >=, <, <=`
  * **논리연산자**: and, or, not

```
x = 123
print(x == 100)
```

```
print((x > 100) and (x < 150))
print((x > 100) and (x < 120))
print((x > 100) or (x < 120))
```

## NoneType

```
x = None   # R에서 NA(Not Available)와 비슷한 역할. is.na()
print(x, type(x))
```

```
# input 함수를 이용해서 정수 2개를 입력받고 저장
# 두 정수의 합(+), 차(-), 곱(*), 나누기결과(/), 나눈 몫(//), 나눈 나머지(%), 거듭제곱(**)
x = int(input('정수 입력>>> '))
y = int(input('정수 입력>>> '))
print(f'{x} + {y} = {x + y}')
print('{} - {} = {}'.format(x, y, x-y))
print('%d X %d = %d'% (x, y, x * y))
print(f'{x} / {y} = {x / y}')
print(f'{x} // {y} = {x // y}')
print(f'{x} % {y} = {x % y}')
print(f'{x}^{y} = {x ** y}')
```

```
print(10 ** -1)
```

```
# 섭씨 온도를 입력받아서, 화씨온도로 변환 후 출력
# F = C * 1.8 + 32

celcius = float(input('섭씨온도 입력: '))
fahrenheit = celcius * 1.8 + 32
print(f'섭씨온도({celcius}) = 화씨온도({fahrenheit})')
```
