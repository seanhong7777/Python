# tuple 

**tuple**

* 원소들을 변경할 수 없는 리스트
* 인덱스를 사용해서 원소를 접근할 수 있음
* append(), insert(), extend(), remove(), pop()와 같은 기능(메서드)들은 제공되지 않음 

```
numbers = (1, 2, 3, 4, 5)
print(numbers)
print(type(numbers))
print(numbers[0])
print(numbers[-1])
```

## tuple slicing 

```
print(numbers[1:3])
```

```
print(numbers[:3])
```

```
print(numbers[3:])
```

```
print(numbers[-1:])  #> slicing의 결과: tuple
```

```
numbers[0] = 100   # 대입 불가 
```

tuple과 for loop

```
print(numbers)
for x in numbers:
    print(x)
```

```
num_list = [x for x in range(10)]  # list comprehension 
print(num_list)
```

```
# empty_tuple = tuple()
empty_tuple = ()
for x in range(10):
    empty_tuple.append(x)  # append 메서드가 없음
print(empty_tuple)
```

## tuple decomposition(분해)

```
one, two = (1, 2) 
print(one)
print(two)
```

```
# list decomposition
gender1, gender2 = ['male', 'female']
print(gender1)
print(gender2)
```

```
numbers = [1, 3, 2, 4, 10]
for index, value in enumerate(numbers):
    print(index, ':', value)
```
