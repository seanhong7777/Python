# 조건문

## else 없는 조건문

```
if 조건식: 
  조건식이 True 일 때 실행할 블록(문장들)
```

* if 조건식 뒤에는 반드시 콜론(:)을 사용'
* if 블록은 반드시 들여쓰기를 해야 함



````
number = -100
if number > 0:
    print(f'{number}는 양수입니다')
    print('if 문 바깥')
````

## else 있는 조건문



```
if 조건식:
    조건식이 True일 때 실행할 블록(문장들)
else:
    조건식이 False일 때 실행할 블록(문장들)
```

* if 조건식과, else 뒤에는 반드시 콜론(:)을 사용 
* if 블록과 else 블록은 반드시 들여쓰기를 해야 함


````
number = 123
if number > 0:
    print(f'{number}는 양수입니다.')
else:
    print(f'{number}는 양수가 아닙니다')
print('if-else 바깥')
````

````
# 정수를 입력받고 변수에 저장 
# 입력받은 정수가 짝수인지, 홀수인지 출력 

number = int(input('정수 입력>>>  '))
if number == 0:
    print(f'0 입니다')
elif number%2 == 0:    
    print(f'{number}은 짝수입니다')
else:
    print(f'{number}는 홀수입니다')
````

## Chained Conditionals (if-elif-else)



```
if 조건식1:
    조건식1이 True인 경우에 실행할 블록
elif 조건식2:
    조건식2이 True인 경우에 실행할 블록
...
else:
    위의 모든 조건식들이 False일때 실행할 블록
```
* elif 조건식은 1개 이상일 수 있음
* else가 필요 없는 경우에는 생략할 수 있음
* 모든 블록들은 들여쓰기가 되어야 함


````
number = 123

if number > 0:
    print('positive')
elif number < 0:
    print('negative')
else:
    print('zero')

print('=== end ===')
````

## 중첩 조건문(Nested Conditionals)

* if 블록, elif 블록, else 블록 안에서 조건문(if, if-else, if-elif-else, ..)이 사용되는 경우
* 중첩 조건문을 사용할 때는 들여쓰기를 주의해야 함. 들여쓰기에 따라서 완전히 다른 로직이 될 수 있음 
* 중첩 조건문은 elif를 사용하거나 논리 연산자(and, or)를 사용해서 중첩되지 않는 조건문으로 만들 수 있는 경우도 있음 

```
if condition1:
    if condition2:
        statements
    else:
        statements

if condition1:
    statements
else:
    if condition2:
        statements
    else:

```



````
number = 1020

if number > 0:
    print('positive number')
else:
    if number < 0:
        print('negative number')
    else:
        print('zero')

print('=== end ===')
````

````
# 국어 점수를 입력받고 변수에 저장
# 영어 점수를 입력받고 변수에 저장
# 수학 점수를 입력받고 변수에 저장
# 세 과목의 평균을 계산하고, 출력
# 평균이 90점 이상이면, A
# 평균이 80점 이상, 90점 미만이면, B
# 평균이 70점 이상, 80점 미만이면, C
# 평균이 70점 미만이면, F 출력
# 세 과목이 점수가 모두 60점이고, 총점이 180

a = int(input('국어 점수>>> '))
b = int(input('영어 점수>>> '))
c = int(input('수학 점수>>> '))

total = a + b + c
avg = total / 3

print(f'총점: {total}, 평균: {avg}')

if avg >= 90: 
    print('A')
elif avg >= 80:
    print('B')
elif avg >= 70:
    print('C')
else:
    print('F')

if total >= 210:
    if a >= 70 and b >= 70 and c >= 70:
        print('Pass')
    else:
        print('Fail')
else: 
    print('Fail')
````
