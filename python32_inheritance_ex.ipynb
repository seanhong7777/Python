```
class Point:
    """2차원 평면 상의 점의 좌표(x, y)를 표현하는 클래스"""
    def __init__(self, x, y):  
        self.x = x
        self.y = y

    def __repr__(self):
        # Point(0, 0) 형태로 출력되도록
        return f'Point({self.x}, {self.y})'

    def move(self, dx, dy):
        """점의 좌표를 x축의 방향으로 dx만큼, y축의 방향으로 dy만큼 이동"""
        self.x += dx
        self.y += dy
    
```

```
pt = Point(0, 0)
print(pt)
print(pt.__repr__())
# print 함수는 객체가 argument로 전달되면 그 객체의 __repr__() 메서드를 호출해 문자열을 리턴받고
# 리턴받은 문자열을 콘솔에 출력해줌
```

```
pt.move(1, 2)
print(pt)
```

```
class Shape:
    """2차원 평면 상의 도형을 표현하는 클래스"""
    def __init__(self, x, y):
        self.point = Point(x, y)

    def __repr__(self):
        return f'Shape[point={self.point}]'
        # return f'shape[point={self.point.__repr__()}]
        # return f'Shape[point=({self.point.x}, {self.point.y})]'

    def move(self, dx, dy):
        """도형을 x축의 방향으로 dx만큼, y축의 방향으로 dy만큼 이동시키는 메서드"""
        self.point.move(dx, dy)
        # self.point.x += dx
        # self.point.y += dy                        

    def area(self):
        raise NotImplementedError('area 메서드가 구현되어야 함!')

    def perimeter(self):
        raise NotImplmentedError('둘레의 길이를 리턴할 수 있는 메서드가 구현되어야 함!')
```

```
sh = Shape(0, 0)
print(sh)
```

```
sh.move(3, 4)
print(sh)
```

```
sh.area()
```

```
class Circle(Shape):   # 1. Shape 클래스를 상속한다
    """중심의 좌표가 (x, y)이고, 반지름의 길이 radius인 원을 표현하는 클래스"""
    
    # 2. __init__ 메서드를 작성
    def __init__(self, x, y, radius):
        super().__init__(x, y)
        self.radius = radius

    # 3. __repr__ 메서드를 작성: Circle(center = Point(0, 0), radius=1)
    def __repr__(self):
        return f'Circle(center = {self.point}, radius={self.radius})'

    # 4. area(), perimeter() 메서드를 override 
    def area(self):
        return math.pi * self.radius ** 2

    def perimeter(self):
        return 2 * math.pi * self.radius   
```

```
c = Circle(0, 0, 1)
print(c)
```

```
c.move(1, 1)  # 부모 클래스인 Shape으로부터 상속받은 move() 메서드 호출
print(c)
```

```
print('넓이:', c.area())
print('둘레:', c.perimeter())
```

```
class Rectangle(Shape):  # 1. Shape 클래스를 상속    
    """꼭지점의 좌표가 (x, y)이고, 가로 길이가 width, 세로 길이가 height인 직사각형을 표현하는 클래스"""
    # 2. __init__ 메서드 작성
    def __init__(self, x, y, width, height):
        super().__init__(x, y)
        self.width = width
        self.height = height

    # 3. __repr__ 메서드 작성: Rectangle(corner=Point(0, 0), width=3, height=4)
    def __repr__(self):
        return f'Rectangle(corner = {self.point}, width = {self.width}, height = {self.height})'

    # 4. area(), perimeter() 메서드를 override
    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width +self.height)
```

```
rect = Rectangle(0, 0, 3, 4)
print(rect)
```

```
rect.move(-1, -2)
print(rect)
```

```
print('넓이:', rect.area())
print('둘레:', rect.perimeter())
```

```
class Square(Rectangle):    # 1. Shape 클래스를 상속
    """꼭지점의 좌표가 (x, y)이고, 한 변의 길이가 length인 정사각형을 표현하는 클래스"""
    # 2. __init__ 메서드 작성
    def __init__(self, x, y, length):   
        super().__init__(x, y, length, length)   # 위의 length가 각 width, height로 들어간다

    def __repr__(self):
        return f'Square(corner={self.point}, length={self.width})'
```

```
sq = Square(0, 0, 3)
print(sq)
```

```
sq.move(-1, -2)
print(sq)
```

```
print('넓이:', sq.area())
print('둘레:', sq.perimeter())
```
