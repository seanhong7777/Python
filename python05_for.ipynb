# 반복문 (Iteration)

**for Statement**

```
for 변수 in iterable-type:
    반복해서 실행할 블록(문장들)
```

* interable-type: 반복 가능한 데이터 타입. range, list, tuple, dict, set, string, ...


````
string = 'Hello'
for ch in string:
    print(ch)
````

````
# range(a, b): [a, b]. a <= x < b 범위의 (1씩 증가하는)정수들
for x in range(1, 6):
    print(x)
````

````
# range(start, stop, step): [start, stop). start >= x < stop 범위의 step 만큼씩 증가하는 정수들
for x in range(1, 10, 2):
    print(x)
````

````
# range(stop): [0, stop). o <= x < stop 범위의 (1씩 증가하는) 정수들
for x in range(5):
    print(x)
````

````
# 아래와 같이 출력되도록 반복문을 작성

for x in range(5, 0, -1):
    print(x)
````

````
# 아래 코드의 pass를 수정해서 10 미만의 짝수들만 출력 
for n in range(10): 
    if n%2 == 0:
        print(n)
````

````
for i in range(0, 6):
    for j in range(i):
        print('*', end ='')
    print()
````

````
for n in range(1, 6):
    print('*' * n)
````

````
for x in (2, 9):
    for y in range(1, 10):
        print(f'{x} x {y} = {x*y}')
    print('...')
````

````
for n in range(1,10):
  print(f'2 x {n} = {2*n}')
````
