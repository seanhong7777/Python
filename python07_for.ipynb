# 중첩 반복문(Nested Interation)

````
for dan in range(2, 10):
    for n in range(1, 10):
        print(f'{dan} X {n} = {dan * n}')
    print('-' * 20)
````

구구단 2단 ~ 9단까지 출력

```
2 x 1 = 2
2 x 2 = 4 
--------------------
3 x 1 =3
3 x 2 = 6
3 x 3 = 9
--------------------
...
```


````
for dan in range(2, 10):
    for n in range(1, 10):
        if n <= dan:
            print(f'{dan} X {n} = {dan * n}')
    print('-' * 20)
````

````
for dan in range(2, 10):
    for n in range(1, dan+1):
        print(f'{dan} X {n} = {dan * n}')
    print('-' * 20)
````

# Break 와 Continue

* Break: iteration을 종료. 반복문 바깥의 코드를 실행. brak가 포함되어 있는 가장 가까운 반복문을 빠져 나오게 됨
* Continue: (반복문 블록 내부 실행문의 수행을 멈추고) 다음 iteration을 수행 

````
for x in range(5):
    if x ==3:
        break
    print(x)
    
````

````
for x in range(5):
    if x ==3:
        continue
    print(x)
````

````
for dan in range(2, 10):
    for n in range(1, 10):
        print(f'{dan} X {n} = {dan * n}')
        if dan == n:
            break
        
    print('-' * 20)
````
