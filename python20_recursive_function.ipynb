**재귀 함수(recursive function)**: 함수 내부에서 자기 자신을 다시 호출하는 함수

```
def factorial(n):
    """
    0! = 1
    1! = 1
    2! = 1 x 2 = 2
    3! = 1 x 2 x 3 
    n! = 1 x 2 x ... x(n-1) x n, n >= 1
    """

    result = None 
    if n == 0:
        result = 1 # 0! = 1
    elif n > 0:
        result = 1
        for x in range(1, n + 1): 
            result *= x

    return result 
```

```
factorial(0) 
```

```
factorial(1)
```

```
factorial(3)
```

```
factorial(-100)
```

```
# 0! = 1
# n! = 1 x 2 x 3 x ... x (n-1) x n = (n-1)! x N, n >= 1
def factorial2(n):
    result = None
    if n == 0:
        result =1
    elif n > 0:
        result = factorial2(n -1) * n

    return result
```

```
for n in range(-1, 6):
    print(f'n={n}! = {factorial2(n)} ')
```

```
def sum_to(n):
    """
    s(0) = 0
    s(1) = 1
    s(2) = 1 + 2
    ...
    s(n) =1 + 2 + 3 + ... + (n-1) + n  = s(n-1) + n, n > 0
    n < 0 경우, None
    """
```

```
def sum_to(n):
    result = None
    if n == 0 :
        result = 0
    elif n > 0:
        result = sum_to(n-1) + n
        
    return result
```

```
sum_to(5)
```

```
for n in range(-1, 6):
    print(f'sum_to({n}) = {sum_to(n)}')
```
