**2선승제 가위/바위/보 게임**

```
def user_choice() -> int: 
    """사용자 입력을 받아서, (1, 2, 3) 중에 하나를 리턴
    """
    while True:
        print('[1] 가위')
        print('[2] 바위')
        print('[3] 보')
        try:
            user = int(input('선택>>> '))
            if user not in (1, 2, 3):
                raise ValueError()    # 에러가 발생되면 except로 간다
        except:
            print('1, 2, 3 중에서만 선택하세요...')
        else:
            return user
            # return 문의 의미: 
            # 1) 함수를 호출한 곳에 값을 반환
            # 2) 함수를 종료 -> break를 따로 하거나, 밖에서 rturn 따로 해줄 필요가 없음
```

```
import random
def computer_choice() -> int:
    """(1, 2, 3) 중에 하나를 랜덤하게 리턴
    """
    return random.randrange(1, 4)
```

```
def show_user_computer_choices(user: int, computer: int) -> None:
    """user, computer 선택을 문자열로 화면에 출력
    """
    print(f'User ({rps[user]}) vs Computer ({rps[computer]})\n')
```

```
def game_result(user: int, computer: int) -> str:
    """user, computer의 가위/바위/보 결과를 문자열('user', 'computer', 'tie'로 리턴)
    """
    game = user - computer
    if game == 0:
        result = 'tie'
        print('무승부\n')
    elif game in (1, -2): 
        result = 'user'
        print('사용자 승\n')
    else:
        result = 'computer'
        print('컴퓨터 승\n')

    return result 
```

```
def end_condition() -> bool:
    """user 또는 computer가 이긴 횟수가 2이면 True, 그렇지 않으면 False를  리턴
    """
    if user_win == required_wins or computer_win == required_wins:
        return True
    else:
        return False
```
