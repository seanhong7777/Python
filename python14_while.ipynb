# while loop(반복문) 



```
[초기화 문장] 
while 조건식: 
    조건식이 참인 경우에 실행할 블록(문장들)
    [조건식의 내용을 변경할 수 있는 문장]
```



````
# 1부터 5까지 출력
n = 1
while n <= 5:
    print(n)
    n += 1
````

````
n = 1
while n <= 5:
    print('*' * n)
    n += 1
````

while loop을 사용해서 구구단 2단을 출력

````
n = 1
while n < 10:
    print(f'2 x {n} = {2 * n}')
    n += 1
````

````
n = 1
while n <= 10:
    if n % 2:
        print(n)
    n += 1
````

## while loop와 break 

````
n = 1
while n <= 5:
    if n == 3:
        break  # break 가 포함된 가장 가까운 반복문(for, while)을 종료

    print(n) 
    n += 1
````

## while loop와 continue

````
n = 1
while n <= 5:
    if n == 3:
        n += 1
        continue

    print(n)
    n += 1
    
````

# 무한 루프(infinite loop)와 break

````
while True:
    msg = input('메시지를 입력하게요. 멈추고 싶으면 Q를 입력하세요>>>  ')
    print(msg)
    if msg == 'Q':
        break
````

````
dan = 2
while dan < 10:
    n = 1
    while dan < 10:
        print(f'{dan} x {n} = {dan * n}')
        if dan == n:
            break 
        n += 1
    print('-' * 20)
    dan += 1
````
