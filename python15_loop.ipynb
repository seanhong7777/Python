1. 아래와 같은 출력을 하는 코드를 작성

```
******
****
***
**
*
```



````
for n in range(5, 0, -1):
    print('*' * n)
````

````
n = 5
while n > 0:
    print('*' * n)
    n -= 1
    if n == 0:
        break
````

2. 주사위 2개를 던졌을때 나오는 눈의 조합(36가지)을 저장하는 리스트를 생성

```
dice = [(1, 1), (1, 2), (1, 3), ..., (6, 4), (6, 5), (6, 6)]
```



````
dice = []
for k in range(1, 7):    
    for v in range(1, 7):
        # eyes = (k, v)
        # dice.append(eyes)
        dice.append((k, v))
print(dice)
````

````
dice = [(x, y) for x in range(1, 7) for y in range(1, 7)]
print(dice)
````

````
dice = []
x = 1
while x < 7: 
    y = 1
    while y < 7: 
        dice.append((x, y))
        y += 1
    x += 1
    
print(dice)
````

3. 주사위 2개를 던졌을때 나오는 눈의 합이 10 이상인 조합들을 저장하는 리스트를 생성

dice_above10 = [(4, 6), (5, 5), (5, 6), (6, 4), (6, 5), (6, 6)]

````
dice_above10 = []
for x in range(1, 7):
    for y in range(1, 7):
        if x + y >= 10:
            dice_above10.append((x, y))
print(dice_above10)
````

````
dice_above10 = []

x = 1
while x < 7:
    y = 1
    while y < 7:
        if x + y >= 10:
            dice_above10.append((x, y))
        y += 1
    x += 1
print(dice_above10)
````

````
dice_above10 = [(x, y) for x in range(1, 7) for y in range(1, 7) if x + y >= 10]
print(dice_above10)
````

4. 1~100까지 합계

```
1 + 2 + 3 + ... + 99 + 100 = ? 
```



````
sum = 0
for n in range(1, 101):
    sum += n 

print(f'총 합계: {sum}')
````

````
sum = 0
n = 1
while n < 101:
    sum += n
    n += 1
print(sum)
````

5. 1~10까지 각 배수의 합계

```
1 + 4 + 9 + 16 + ... + 81 + 100 = ? 
```



````
sum = 0
for n in range(1, 11):
    sum += n ** 2

print(f'총 합계: {sum}')
````

````
sum = 0
n = 1
while n < 11:
    sum += n ** 2
    n += 1
print(sum)
````

6. 아래 조건을 만족하는 가장 큰 정수 

```
1 + 2 + 3 + ... + (n - 1) + n <= 100 를 만족하는 가장 큰 정수 n? 
```


````
sum = 0 
n = 1

while True:
    sum += n
    if sum >= 100:
        break    
    n += 1

print(f'가장 큰 정수 n : {n}')
````
