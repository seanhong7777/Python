**여러가지 print 방법**

```
print("Hello, Python!")
```

print()  함수에 argument를 여러개 전달하면, 각각의 argument를 공백으로 연결해서 한 줄에 출력해줌

```
print('안녕', '여러분', 'hello')
```

```
print(1, 2)
```

```
print(1 + 2)
```

```
print(1, '+', 2, '=', (1+2))
```

```
# Python 주석
# #으로 시작해서 그 줄(line)의 끝까지 

# 변수(variable): 값(자료)를 저장한 곳을 찾아갈 수 있는 이름
# 변수이름 = 값 
name = '오쌤'
age = 16

print(name)
print(age)
```

```
print('이름:', name, '나이:', age)
```

```
print("I'm a boy.")
```

```
print('I\'m a boy.')  # escape character \
```

```
print('이름:', name, ',', '나이:', age)
```

```
print('이름: %s, 나이: %d' % (name, age))
# %s: 문자열
# %d: 정수(integer)
# %f: 실수. 부동소수점(floating point number)
```

```
# format() 메서드 이용
print('이름: {}, 나이: {}'.format(name, age))  # placeholder {}
```

```
print('{} x {} = {}'.format(2, 3, 2*3))
```

```
# Python 3.6 이상 버전부터 새로 추가된 print 방법 
print(f'이름: {name}, 나이: {age}')
```

```
print(f'나이: {age +1}')
```

```
# 사용자 입력(키보드로부터 문자열 입력) 
name = input('이름을 입력하세요>>> ')
print(f'안녕하세요, {name}!')
```
