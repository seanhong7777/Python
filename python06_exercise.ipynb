# for-in 구문 연습

````
# end 파라미터: 프린트가 끝난 후 어떤 동작을 할 것인지를 설정.
# end='\n': 줄바꿈(enter). 기본값.
print('hello', end=' ')
print('안녕하세요', end=' ')
print('Hello Python!', end=' ')
````

````
# sep 파라미터: 출력할 argument들을 어떤 문자(들)로 구분(separate)할지를 설정.
# sep=' ': 공백(space 1개). 기본값
print('hello', '안녕하세요', 'Python', sep='::')
````

## 동요를 불러볼까요?

아래와 같이 출력되도록 코드를 작성

```
1 little 2 little 3 little indian
4 little 5 little 6 little indian
7 little 8 little 9 little indian
10 little indian boys!
```


````
for n in range(1, 11):
    print(n, 'little', end = ' ')
    if n%3 == 0: 
        print('indian')
    
print('indian boys!')
````

````
for i in range(1, 10, 3):
    print(f'{i} little {i+1} little {i+2} little indian')
print('10 little indian boys!')
````

## 369 게임을 해볼까요?

아래와 같이 출력되도록 코드 작성

```
1  2  *  4  5  *  7  8  *  10
11 12 *  14 15 *  17 18 *  20
21 22 *  24 25 *  27 28 *  *
*  *  *  *  *  *  *  *  *  40
...
81 82 *  84 85 *  87 88 *  * 
*  *  *  *  *  *  *  *  *  100
```


````
for n in range(1, 101):
    n_1 = n % 10     # n을 10으로 나눈 나머지 -> 숫자 n의 1의 자릿수
    n_10 = n // 10   # n을 10으로 나눈 몫 -> 숫자 n의 10의 자릿수

    # cond_n1 = (n_1 == 3) or (n_1 == 6) or (n_1 ==9)
    cond_n1 = n_1 in (3, 6, 9)
    cond_n10 = n_10 in (3, 6, 9)

    if cond_n1 and cond_n10:
        print('**', end = '\t')
    elif cond_n1 or cond_n10: 
        print('*', end = '\t')
    else:
        print(n, end = '\t')

    if n % 10 == 0:
        print()
````

````
for n in range(1, 101):
    n_str = str(n)
    if ('3' in n_str) or ('6' in n_str) or ('9' in n_str):
        print('*', end = '\t')
    else: 
        print(n_str, end = '\t')

    if n % 10 == 0:
        print()
````

````
for i in range(1, 101):
    if ('3' in str(i)) or ('6' in str(i)) or ('9' in str(i)):
        print(str('*'), end = '\t')
    else: 
        print(str(i), end = '\t')
    if i%10 == 0:
        print()
````

## 삼각형을 그려볼까요?

아래와 같이 출력되도록 코드 작성

```
*
**
***
****
*****
****
***
**
*
```


````
for n in range(1, 6):
    print('*' * n, end = '\n')
for n in range(4, 0, -1):
    print('*' * n, end = '\n')
````

````
for n in range(1, 10):
    if n <= 5:
        print('*' * n)
    else:
        print('*' * (10 - n))
````

````
for x in range(-4, 5):
    print('*' * (5 - abs(x)))
````

````
for i in range(6):
    print('*' * i, end = '\n')
for j in range(5, 0, -1):
    print('*' * (j-1), end = '\n')
````
